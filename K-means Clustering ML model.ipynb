```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
#read dataset
df = pd.read_csv("income.csv");
df.head()
```

```python
df
```

```python
#plot the data
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
#initialize clusters
km = KMeans(3)
```

```python
#fit the model and predict together
y_predicted = km.fit_predict(df[['Age','Income($)']])
```

```python
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df
```

```python
#plot the data
plt.scatter(df['Age'],df['Income($)'],c=df['cluster'],cmap="rainbow")
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
sse=[]
```

```python
k_rng = range(1,10) #iterate from 1 to 9
```

```python
for k in k_rng:
    #in each iteration, create a new model with clusters =k
    km = KMeans(n_clusters=k)
    #create a new dara frame excluding the name col rather than passing the cloumn names
    #km.fit(<dataframe>
    km.fit(df[['Age',"Income($)"]])
    #inertia will give the sse
    sse.append(km.inertia_)
```

```python
plt.xlabel('K') 
plt.ylabel('Sum of squared error') 
plt.plot(k_rng,sse) 
```
